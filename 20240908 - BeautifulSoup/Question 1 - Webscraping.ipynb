```python
import re
```

```python
pattern = "^a{1-6})$"
```

```python
tag = ['h1', 'h2', 'h3', 'h4', 'h5', 'h6']
```

```python
for tag in tag:
    match = re.match(pattern, tag)
    if match:
        (print) tag + " " match the pattern.

    else:
        print(tag + " not match the pattern.")
```
